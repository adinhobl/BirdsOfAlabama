# View Data

```python
np.random.seed(1)
data = ImageDataBunch.from_folder(path=data_path, valid_pct=0.2, 
                                  ds_tfms=get_transforms(), size=224, num_workers=4).normalize(imagenet_stats)
```

```python
# If you already cleaned your data, run this cell instead of the one before
# np.random.seed(1)
# data = ImageDataBunch.from_csv(".", folder=".", valid_pct=0.2, csv_labels='cleaned.csv',
#                                 ds_tfms=get_transforms(), size=224, num_workers=4).normalize(imagenet_stats)
```

```python
data.c
```

```python
data.show_batch(rows=3)
```

```python
print(" Training dataset size:", len(data.train_ds), '\n', "Validation dataset size:", len(data.valid_ds))
```

# Create CNN & Begin Training

```python
learn = create_cnn(data, models.resnet34, metrics=accuracy)
```

```python
learn.fit_one_cycle(4)
```

```python
learn.save('stage-1')
```

```python
learn.unfreeze()
```

```python
learn.lr_find()
```

```python
learn.recorder.plot()
```

# Continue Training

```python
learn.freeze()
```

```python
learn.fit_one_cycle(2, max_lr=slice(1e-5, 1e-4))
```

```python
learn.fit_one_cycle(5, max_lr=slice(3e-4, 1e-4))
```

```python
learn.save('stage-2')
```

It seems like we are getting a little closer to having a good model. Time to check what our bad actors are.

# Interpretation

```python
interp = ClassificationInterpretation.from_learner(learn)
```

```python
interp.plot_confusion_matrix(figsize=(25,25))
```
